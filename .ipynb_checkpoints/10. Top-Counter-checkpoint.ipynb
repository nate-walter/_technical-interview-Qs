We have a list with identifiers of form “id-SITE”. Show the top 3 sites. You can break ties in any way you want.



```python
sites = ['abcdef-ID', 'ghijkl-IN', 'mnopqrs-IN', 'tuvwxyz-PL', 
         'efptozl-PL', 'pedpecf-PL', 'dedfczpf-UA']
```

```python
def counter(lst):
    ans = {}
    for i in lst:
        site = i[-2:]
        ans[site] = ans.get(site, 0 ) + 1
    return ans
    


def top_counter(lst):
    site_dict = counter(lst) # uses Question 8 function
    top_sites = sorted(site_dict, reverse=True, key=site_dict.get)[:3]
    return{key: site_dict[key] for key in top_sites}
    
```
